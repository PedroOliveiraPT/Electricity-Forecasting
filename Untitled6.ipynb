```python
from influxdb import InfluxDBClient
```

```python
db_client = InfluxDBClient(host="127.0.0.1",port=8086, username="atnoguser", password="atnogpass", database='bosch')
query_api = db_client.get_list_measurements()
query_api
```

```python
query = 'Select last(*) from S_SUM'
result = db_client.query(query)
cpu_points = list(result.get_points(measurement='S_SUM'))
cpu_points
```

```python
import pandas as pd

query = 'Select * from '
measurement = query_api[0]['name']
result = db_client.query(query + measurement)
cpu_points = list(result.get_points(measurement=measurement))
ds = {'ts': [], 'AE_L1': []}
for res in cpu_points:
    ds['ts'].append(res['time'])
    ds[measurement].append(res['value'])

for i in query_api[1:-1]:
    measurement = i['name']
    result = db_client.query(query + measurement)
    cpu_points = list(result.get_points(measurement=measurement))
    ds[measurement] = [res['value'] for res in cpu_points]
```

```python
query = 'Select * from '
measurement = 'l1_n'
result = db_client.query(query + measurement)
cpu_points = list(result.get_points(measurement=measurement))
print(cpu_points)
```

```python
del ds['l1_n']
for i in ds:
    print(i, len(ds[i]))
```

```python
df = pd.DataFrame(ds)
df.head()
df.to_csv('data/influx_data.csv')
```
